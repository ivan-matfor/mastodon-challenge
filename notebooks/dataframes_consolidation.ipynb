## Data Availability

This repository does not provide raw datasets.

To reproduce the analysis:

1. Run the extraction notebook (`pulling_data_API.ipynb`) for the desired hashtags.
2. Save each dataset as a separate CSV file in `data/raw/`.
3. Ensure identical `DAYS_BACK` configuration across tags to preserve temporal comparability.
4. Then execute this consolidation notebook.

The analysis is designed to work on reproducible API snapshots rather than static distributed datasets.
